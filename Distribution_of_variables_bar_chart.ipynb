```python
import pandas as pd
import numpy as nmp
import matplotlib.pyplot as plt
```

### **Reading the dataset**

```python
df = pd.read_csv("Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_5871594.csv")
```

###**Plotting the Bar Chart**

```python
population_count = df['Region'].value_counts()
bar_width = 1
x = range(len(population_count.index))
```

```python
plt.bar(population_count.index, population_count.values, color = "orange")
plt.xlabel("Region", weight = "bold", fontsize = '14')
plt.ylabel("Count", weight = "bold", fontsize = '14')
plt.title("Distribution of Region", color = "#4C0033", weight = "bold", fontsize = '20')

plt.xticks(x, population_count.index, rotation = 60)
```
